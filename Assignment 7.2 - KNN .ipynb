# KNN for Regression

```python
import pandas as pd 
df = pd.read_csv('weight-height.csv')
```

```python
df2 = df.copy() 
```

```python
df2
```

# Encoding
## Manually Encoded

Gender: Male -> 1
        Female -> 0

```python
df2 = df.replace({'Male':1,'Female':0})
```

```python
df2
```

```python
y = df2['Weight']
y
```

# Train_Test_split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(df2[['Gender','Height']],y, test_size=30, random_state=42)
```

```python
xtrain
```

```python
ytrain
```

```python
xtrain
```

```python
xtrain
```

# Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(xtrain, ytrain) #training
```

```python
reg.predict(xtest)
```

# Evaluate the Model (Testing and training Accuracy, MSE for testing)

```python
reg.score(xtest,ytest)
```

```python
reg.score(xtrain,ytrain)
```

```python
#mse
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ytest, reg.predict(xtest) )
```

```python
mse
```


# Apply KNN Regressor:

```python
from sklearn.neighbors import KNeighborsRegressor
knn_reg = KNeighborsRegressor(n_neighbors=3)
```

```python
knn_reg.fit(xtrain,ytrain)
```

```python
knn_reg.predict(xtest)
```

# Evaluate the Model (Testing and training Accuracy, MSE for testing)

```python
knn_reg.score(xtest, ytest)
```

```python
knn_reg.score(xtrain, ytrain)
```

```python
#mse2
from sklearn.metrics import mean_squared_error
mse2 = mean_squared_error(ytest, knn_reg.predict(xtest) )
```

```python
mse2
```

```python
print('As we can see mse of linear regression is {} and mse of KNN Regression is {}'.format(mse,mse2))
print('less cost is always better so linear Regression is better than KNN Regression')
```
